```python
import ast

def clean_crew(text):

    names = []

    try:

        crew_list = ast.literal_eval(text)

        for item in crew_list:

            # خد الـ Director فقط
            if item['job'] == 'Director':

                names.append(
                    item['name']
                )

    except:
        pass

    return " ".join(names)
```

```python
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(
    stopwords.words('english')
)

lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(text):

    text = str(text).lower()

    # remove special chars
    text = re.sub(
        r'[^a-zA-Z0-9 ]',
        ' ',
        text
    )

    # tokenize
    words = text.split()

    # remove stopwords
    words = [
        w for w in words
        if w not in stop_words
    ]

    # lemmatization
    words = [
        lemmatizer.lemmatize(w)
        for w in words
    ]

    return " ".join(words)
```

```python
def preprocess_title(text):

    text = str(text).lower()

    # remove special chars only
    text = re.sub(
        r'[^a-zA-Z0-9 ]',
        ' ',
        text
    )

    return text
```

```python
movies['clean_overview'] = movies[
    'overview'
].apply(preprocess)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
movies['clean_cast'] = movies[
    'main_cast'
].apply(preprocess_title)
```

```python
movies['clean_director'] = movies[
    'director'
].apply(preprocess_title)
```

```python
title_vectorizer = TfidfVectorizer(
    ngram_range=(1,2)
)

title_matrix = title_vectorizer.fit_transform(
    movies['clean_title']
)
```

```python
overview_vectorizer = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1,2)
)

overview_matrix = overview_vectorizer.fit_transform(
    movies['clean_overview']
)
```

```python
cast_vectorizer = TfidfVectorizer(
    ngram_range=(1,2)
)

cast_matrix = cast_vectorizer.fit_transform(
    movies['clean_cast']
)
```

```python
director_vectorizer = TfidfVectorizer(
    ngram_range=(1,2)
)

director_matrix = director_vectorizer.fit_transform(
    movies['clean_director']
)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def search_movies(
    query,
    title_weight=2.0,
    overview_weight=1.0,
    cast_weight=1.5,
    director_weight=1.5,
    top_k=10
):

    # =========================
    # preprocess query
    # =========================

    clean_query_title = preprocess_title(query)

    clean_query_overview = preprocess(query)

    # =========================
    # title similarity
    # =========================

    title_query_vector = title_vectorizer.transform(
        [clean_query_title]
    )

    title_scores = cosine_similarity(
        title_query_vector,
        title_matrix
    ).flatten()

    # =========================
    # overview similarity
    # =========================

    overview_query_vector = overview_vectorizer.transform(
        [clean_query_overview]
    )

    overview_scores = cosine_similarity(
        overview_query_vector,
        overview_matrix
    ).flatten()

    # =========================
    # cast similarity
    # =========================

    cast_query_vector = cast_vectorizer.transform(
        [clean_query_title]
    )

    cast_scores = cosine_similarity(
        cast_query_vector,
        cast_matrix
    ).flatten()

    # =========================
    # director similarity
    # =========================

    director_query_vector = director_vectorizer.transform(
        [clean_query_title]
    )

    director_scores = cosine_similarity(
        director_query_vector,
        director_matrix
    ).flatten()

    # =========================
    # final weighted score
    # =========================

    final_scores = (
        title_weight * title_scores
        +
        overview_weight * overview_scores
        +
        cast_weight * cast_scores
        +
        director_weight * director_scores
    )

    # =========================
    # top results
    # =========================

    top_indices = final_scores.argsort()[-top_k:][::-1]

    results = movies.iloc[top_indices][[
        'title',
        'genres_clean',
        'main_cast',
        'director',
        'vote_average',
        'popularity'
    ]]

    results['score'] = final_scores[top_indices]

    return results
```

```python
search_movies(
    "spider",
    cast_weight=5.0
)
```

```python
from rank_bm25 import BM25Okapi
```

```python
movies['bm25_text'] = (
    movies['clean_title'] + ' ' +
    movies['clean_overview'] + ' ' +
    movies['clean_cast'] + ' ' +
    movies['clean_director']
)
```

```python
tokenized_corpus = [
    doc.split()
    for doc in movies['bm25_text']
]
```

```python
bm25 = BM25Okapi(
    tokenized_corpus
)
```

```python
def bm25_search(
    query,
    top_k=10
):

    # preprocess query
    clean_query = preprocess(query)

    # tokenize query
    tokenized_query = clean_query.split()

    # BM25 scores
    scores = bm25.get_scores(
        tokenized_query
    )

    # top results
    top_indices = scores.argsort()[-top_k:][::-1]

    results = movies.iloc[top_indices][[
        'title',
        'genres_clean',
        'main_cast',
        'director',
        'vote_average',
        'popularity'
    ]]

    results['score'] = scores[top_indices]
    results = results.sort_values(
    by=['score', 'popularity'],
    ascending=False
)

    return results
```

```python
bm25_search(
    "dwayne johnson"
)
```

#Evaluation

```python
test_queries = {
    "brad pitt": [
        "Fight Club",
        "Se7en",
        "World War Z"
    ],

    "batman": [
        "Batman Begins",
        "The Dark Knight",
        "The Dark Knight Rises"
    ],

    "spider man": [
        "Spider-Man",
        "Spider-Man 2",
        "Spider-Man 3"
    ]
}
```

```python
def precision_at_k(
    retrieved_titles,
    relevant_titles,
    k=10
):

    retrieved_k = retrieved_titles[:k]

    relevant_found = 0

    for movie in retrieved_k:

        if movie in relevant_titles:

            relevant_found += 1

    return relevant_found / k
```

```python
for query, relevant_movies in test_queries.items():

    results = bm25_search(query)

    retrieved_titles = results[
        'title'
    ].tolist()

    precision = precision_at_k(
        retrieved_titles,
        relevant_movies,
        k=10
    )

    print(f"Query: {query}")

    print(f"Precision@10: {precision}")

    print("-" * 40)
```
